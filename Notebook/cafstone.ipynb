```python
import pandas as pd
from google.colab import files
from tabulate import tabulate
```

```python
url = 'https://raw.githubusercontent.com/Sctian514/cafstone/main/Restomedan3.csv'
df = pd.read_csv(url)
display(df)
```

```python
df['reservable'] = df['reservable'].replace('unknown', 'False')
df['outdoorSeating'] = df['outdoorSeating'].replace('unknown', 'False')
df['servesDessert'] = df['servesDessert'].replace('unknown', 'False')
df['goodForChildren'] = df['goodForChildren'].replace('unknown', 'False')
df['priceLevel'] = df['priceLevel'].replace('unknown', 'PRICE_LEVEL_UNKNOWN')
df['servesBeer'] = df['servesBeer'].replace('unknown', 'False')
df['servesWine'] = df['servesWine'].replace('unknown', 'False')
df['liveMusic'] = df['liveMusic'].replace('unknown', 'False')
df['goodForGroups'] = df['goodForGroups'].replace('unknown', 'False')
df['acceptsCreditCards'] = df['acceptsCreditCards'].replace('unknown', 'False')
df['acceptsDebitCards'] = df['acceptsDebitCards'].replace('unknown', 'False')
df['acceptsCashOnly'] = df['acceptsCashOnly'].replace('unknown', 'False')
df['acceptsNfc'] = df['acceptsNfc'].replace('unknown', 'False')
df['servesCocktails'] = df['servesCocktails'].replace('unknown', 'False')
df['rating'] = df['rating'].replace('unknown', '4,4')

display(df)
```

```python
duplicate_entries = df[df.duplicated(subset=['id','displayName', 'formattedAddress'], keep=False)]

duplicate_count = duplicate_entries.groupby(['id','displayName', 'formattedAddress']).size().reset_index(name='count')

duplicate_count = duplicate_count[duplicate_count['count'] > 1]

duplicate_count
```

```python
num_duplicates = df.duplicated(subset=['id'], keep=False).sum()
print(f"\nJumlah baris duplikat (berdasarkan ID): {num_duplicates}")
```

```python
df_unique = df.drop_duplicates(subset=['id'])
print("\nDataframe setelah menghapus duplikat:")
display(df_unique)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
import numpy as np

# Memuat data dari CSV
df_new = pd.read_csv('/content/resto_medan2.csv')

# Tampilkan beberapa baris pertama dari data untuk verifikasi
display(df_new.head())
```

```python
df_new.info()
```

```python
# Define feature columns for preferences and type
preference_features = ['servesBeer', 'servesWine', 'servesCocktails', 'goodForChildren',
                       'goodForGroups', 'reservable', 'outdoorSeating', 'liveMusic', 'servesDessert',
                       'priceLevel','acceptsCreditCards', 'acceptsDebitCards', 'acceptsCashOnly', 'acceptsNfc'
                       ]

type_feature = 'Type_encoded'

# Encode 'priceLevel' and 'Type' column
label_encoder_priceLevel = LabelEncoder()
df_new['priceLevel'] = label_encoder_priceLevel.fit_transform(df_new['priceLevel'])

label_encoder_type = LabelEncoder()
df_new['Type_encoded'] = label_encoder_type.fit_transform(df_new['Type'])

# Normalize preference features
scaler = StandardScaler()
X_preferences = df_new[preference_features].values
X_preferences_scaled = scaler.fit_transform(X_preferences)

# Extract type features
X_type = df_new[type_feature].values

# Split data into training and testing sets for preferences
X_train_prefs, X_test_prefs = train_test_split(X_preferences_scaled, test_size=0.2, random_state=42)
```

```python
# Define model architecture for type
input_layer_type = Input(shape=(1,))
dense_layer_1_type = Dense(64, activation='relu')(input_layer_type)
dense_layer_2_type = Dense(32, activation='relu')(dense_layer_1_type)
output_layer_type = Dense(1, activation='linear')(dense_layer_2_type)

# Build model for type
model_type = Model(inputs=input_layer_type, outputs=output_layer_type)
model_type.compile(optimizer='adam', loss='mse')

# Train model for type
X_train_type, X_test_type = train_test_split(X_type, test_size=0.2, random_state=42)
model_type.fit(X_train_type.reshape(-1, 1), X_train_type, epochs=50, batch_size=8, validation_data=(X_test_type.reshape(-1, 1), X_test_type))
```

```python
# Define model architecture for preferences
input_layer_prefs = Input(shape=(len(preference_features),))
dense_layer_1_prefs = Dense(64, activation='relu')(input_layer_prefs)
dense_layer_2_prefs = Dense(32, activation='relu')(dense_layer_1_prefs)
output_layer_prefs = Dense(len(preference_features), activation='linear')(dense_layer_2_prefs)

# Build model for preferences
model_prefs = Model(inputs=input_layer_prefs, outputs=output_layer_prefs)
model_prefs.compile(optimizer='adam', loss='mse')

# Train model for preferences
model_prefs.fit(X_train_prefs, X_train_prefs, epochs=50, batch_size=8, validation_data=(X_test_prefs, X_test_prefs))
```

```python
# Function to recommend cafes based on preferences
def recommend_cafes_by_preferences(user_preferences, model_prefs, data, scaler):
    if user_preferences is not None:
        user_preferences_scaled = scaler.transform(user_preferences)
        preference_predictions = model_prefs.predict(user_preferences_scaled)
        preference_distances = np.linalg.norm(X_preferences_scaled - preference_predictions, axis=1)
        similar_cafes_indices = preference_distances.argsort()[:10]
        return data.iloc[similar_cafes_indices]
    else:
        raise ValueError("User preferences must be provided")

# Function to recommend cafes based on type
def recommend_cafes_by_type(user_type, model_type, data, X_type):
    type_predictions = model_type.predict(np.array(user_type).reshape(-1, 1))
    type_distances = np.abs(X_type - type_predictions.flatten())
    similar_cafes_indices = type_distances.argsort()[:10]
    return data.iloc[similar_cafes_indices]
```

```python
# Example of recommendations based on preferences
test_user_preferences = [
    [0, 0,0, 1, 1, 1, 1, 1, 1,2, 1, 1, 1, 1]
]

# Get recommendations based on preferences
test_recommendations_by_preferences = recommend_cafes_by_preferences(np.array(test_user_preferences), model_prefs, df_new, scaler)

# Show recommendations
print(f"\nRecommendations for user preferences {test_user_preferences[0]}:")
display(test_recommendations_by_preferences)
```

```python
# Example of recommendations based on type
test_user_type = [13]

# Get recommendations by type
test_recommendations_by_type = recommend_cafes_by_type(test_user_type, model_type, df_new, X_type)

# Show recommendations
print(f"\nRecommendations for type {test_user_type[0]} only:")
display(test_recommendations_by_type)
```
